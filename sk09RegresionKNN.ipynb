### Regresion con K-vecinos más cercanos
#### Dataset: Diabetes

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = load_diabetes()
```

```python
print(data.data.shape)
print(data.target.shape)
```

```python
data.feature_names
```

```python
print(data.data)
```

```python
data.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.data, data.target)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf = KNeighborsRegressor()
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
df_y_test = pd.DataFrame(y_test, columns=['y_test'])
df_y_pred = pd.DataFrame(y_pred, columns=['y_pred'])
pd.concat([df_y_test, df_y_pred], axis=1)
```

```python
plt.plot(y_test)
plt.plot(y_pred)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y_test', 'predicciones'])
plt.show()
```

```python
mean_squared_error(y_test, y_pred)
```

$ MSE = \frac{1}{n} \sum_{i=1} ^ {n} (t_i - y_i)^{2} $

$ MSE = \frac{1}{n} \sum_{i=1} ^ {n} (ytest_i - ypred_i)^{2} $
